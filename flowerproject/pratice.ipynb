```python
import random
from decimal import Decimal
from django.contrib.auth.models import User
from flowerapp.models import Category, Flower, Customer, Order, OrderItem
from django.db import transaction
```

```python
categories = [
    Category(
        name=f"Category {i}",
        descrition=f"Description for category {i}"
    )
    for i in range(1, 11)
]

Category.objects.bulk_create(categories)

print("Categories created:", Category.objects.count())
```

```python
categories = list(Category.objects.all())

flowers = []

for i in range(51):
    flowers.append(
        Flower(
            name=f"Flower {i}",
            description=f"Beautiful flower number {i}",
            price=Decimal(random.randint(100, 1000)),
            category=random.choice(categories)
        )
    )

Flower.objects.bulk_create(flowers)

print("Flowers created:", Flower.objects.count())
```

```python
users = []
customers = []

for i in range(21):
    users.append(
        User(
            username=f"user{i}",
            email=f"user{i}@example.com"
        )
    )

User.objects.bulk_create(users)

created_users = list(User.objects.filter(username__startswith="user"))

for user in created_users:
    customers.append(
        Customer(
            user=user,
            phone_number=f"99999{random.randint(10000,99999)}",
            address="Sample address"
        )
    )

Customer.objects.bulk_create(customers)

print("Customers created:", Customer.objects.count())
```

```python
orders = list(Order.objects.all())
flowers = list(Flower.objects.all())

items = []

for i in range(101):
    flower = random.choice(flowers)

    items.append(
        OrderItem(
            order=random.choice(orders),
            flower=flower,
            quantity=random.randint(1, 5),
            unit_price=flower.price
        )
    )

OrderItem.objects.bulk_create(items)

print("OrderItems created:", OrderItem.objects.count())
```
